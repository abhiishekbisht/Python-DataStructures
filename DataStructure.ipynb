```python
# Dictionary by input - > 

n = int(input("No. of key-value pairs"))
dict = {}
for i in range(n):
    key = input(f"Enter Key {i+1}")
    value = input(f"Enter Value : ")
    dict[key] = value
key = list(dict.keys())
print(key[1])
print(dict[key[0]])
```
